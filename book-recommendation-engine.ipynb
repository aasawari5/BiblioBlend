```python
#top 10 books with highest rating
top_ten=df[df['ratings_count']>1000000]
top_ten=top_ten.sort_values(by='ratings_count', ascending=False).head(10)
top_ten_rating=top_ten.sort_values(by='average_rating', ascending=False)
top_ten_rating
```

```python
#visualize average ratings of top 10 books
sns.set_style('whitegrid')
plt.figure(figsize=(4, 4))
plt.barh(top_ten_rating['title'], top_ten_rating['average_rating'], color='green')
plt.xlabel('Average Rating')
plt.ylabel('Title')
plt.title('Top Ten Rating by Title')
plt.show()
```

```python
#count the number of titles for each author, and then display the top ten authors with the highest number of titles
book_aut=df.groupby('authors')['title'].count().reset_index().sort_values('title', ascending=False).head(10).set_index('authors')
print(book_aut)
```

```python
#distribution of book count among top 10 authors
plt.figure(figsize=(4, 4))
plt.pie(book_aut['title'], labels=book_aut.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 10 Authors by Number of Books')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# number of books by each author
colors = plt.cm.tab10(np.arange(len(book_aut)))
plt.figure(figsize=(4, 4))
book_aut['title'].plot(kind='bar', color=colors)
plt.title('Top 10 Authors by Number of Books')
plt.xlabel('Authors')
plt.ylabel('Number of Books')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Group by 'publisher' and count the number of books
publisher_counts = df.groupby('publisher')['title'].count().reset_index()
# Sort the publishers by the number of books in descending order
publisher_counts = publisher_counts.sort_values('title', ascending=False).head(10)
print(publisher_counts)
```

```python
# Create a pie chart
# distribution of books among top 10 publishers
top_publishers = publisher_counts.head(10)

plt.figure(figsize=(4, 4))
plt.pie(top_publishers['title'], labels=top_publishers['publisher'], autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Publishers by Number of Books')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
#number of books by each publisher in bar chart visualization
top_publishers = publisher_counts.head(10)
num_publishers = len(top_publishers)
colors = plt.cm.tab10(np.linspace(0, 1, num_publishers))

plt.figure(figsize=(4, 4))

# Create a bar plot using the custom color palette
plt.bar(top_publishers['publisher'], top_publishers['title'], color=colors)

plt.title('Top 10 Publishers by Number of Books')
plt.xlabel('Publishers')
plt.ylabel('Number of Books')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

3. DATA PREPARATION

```python
df2=df.copy()
```

```python
df2.loc[(df2['average_rating']>=0)&(df2['average_rating']<=1),'rating between']="between 0 and 1"
df2.loc[(df2['average_rating']>1)&(df2['average_rating']<=2),'rating between']="between 1 and 2"
df2.loc[(df2['average_rating']>2)&(df2['average_rating']<=3),'rating between']="between 2 and 3"
df2.loc[(df2['average_rating']>3)&(df2['average_rating']<=4),'rating between']="between 3 and 4"
df2.loc[(df2['average_rating']>4)&(df2['average_rating']<=5),'rating between']="between 4 and 5"
```

```python
df2.sort_values('average_rating')
```

```python
rating_df=pd.get_dummies(df2['rating between'])
language_df=pd.get_dummies(df2['language_code'])
```

```python
language_df
```

```python
rating_df
```

```python
feature=pd.concat([rating_df, language_df, df2['average_rating'], df2['ratings_count']], axis=1)
```

```python
feature 
```

4. MODEL TRAINING (KNN)

```python
from sklearn.preprocessing import MinMaxScaler

minmax=MinMaxScaler()

feature=minmax.fit_transform(feature)
```

```python
feature
```

```python
from sklearn import neighbors
from sklearn.model_selection import train_test_split

model=neighbors.NearestNeighbors(n_neighbors=8, algorithm='ball_tree')

model.fit(feature)

dist, idlist=model.kneighbors(feature)
```
